```python
def plot_degree_dist(J):
    all_degrees = list(dict(nx.degree(J)).values())
    unique_degrees = list(set(list(dict(nx.degree(J)).values())))
    
    count_of_degrees = []
    for i in unique_degrees:
        x = all_degrees.count(i)
        count_of_degrees.append(x)
    plt.plot(unique_degrees, count_of_degrees)
    plt.show()
plot_degree_dist(J)
```

```python
def plot_degree_dist(J):
    all_degrees = list(dict(nx.degree(J)).values())
    unique_degrees = list(set(list(dict(nx.degree(J)).values())))
    
    count_of_degrees = []
    for i in unique_degrees:
        x = all_degrees.count(i)
        count_of_degrees.append(x)
    plt.plot(unique_degrees, count_of_degrees, 'yo-')
    plt.xlabel('Degrees')
    plt.ylabel('Number of nodes')
    plt.title('Degree distribution of karate network')
    plt.show()
plot_degree_dist(J)
```

```python
def plot_degree_dist(J):
    all_degrees = list(dict(nx.degree(J)).values())
    unique_degrees = list(set(list(dict(nx.degree(J)).values())))
    
    count_of_degrees = []
    for i in unique_degrees:
        x = all_degrees.count(i)
        count_of_degrees.append(x)
    plt.loglog(unique_degrees, count_of_degrees, 'yo-')
    plt.xlabel('Degrees')
    plt.ylabel('Number of nodes')
    plt.title('Degree distribution of karate network')
    plt.show()
plot_degree_dist(J)
```
