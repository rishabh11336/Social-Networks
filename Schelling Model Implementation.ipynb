```python
import networkx as nx
```

```python
N = 10
```

```python
G = nx.grid_2d_graph(N,N)
```

```python
import matplotlib.pyplot as plt
```

```python
nx.draw(G)
```

```python
list(G.nodes())
```

```python
pos = dict((n,n) for n in list(G.nodes()))
```

```python
nx.draw(G, pos)
```

```python
nx.draw(G, pos, with_labels=1)
```

```python
labels = dict(((i,j), i*10+j) for i,j in list(G.nodes()))
```

```python
nx.draw(G, pos, with_labels = False)
```

```python
nx.draw_networkx_labels(G, pos, labels = labels)
```

```python
plt.show()
```

```python
nx.draw_networkx_labels(G, pos, labels = labels)
```

```python
nx.draw_networkx_labels(G, pos, labels = labels)
plt.show()
```

```python
labels = dict(((i,j), i*10+j) for i,j in list(G.nodes()))
```

```python
nx.draw_networkx_labels(G, pos, labels = labels)
plt.savefig('2D-grid1.jpg')
plt.clf()
plt.cla()
```

```python
for (u,v) in list(G.nodes()):
    if (u+1 <= N-1) and (v-1>=0):
        G.add_edge((u,v),(u+1,v-1))
```

```python
nx.draw(G, pos)
```

```python
for (u,v) in list(G.nodes()):
    if (u+1 <= N-1) and (v+1 <= N-1):
        G.add_edge((u,v),(u+1,v+1))
```

```python
nx.draw(G, pos)
```

```python
nx.draw_networkx_labels(G, pos, labels = labels)
plt.show()
```
