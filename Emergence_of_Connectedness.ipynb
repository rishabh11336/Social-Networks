```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy
```

```python
#Add n number of nodes in ghraph
def add_nodes(n):
    G = nx.Graph()
    G.add_nodes_from(range(n))
    return G
```

```python
G = add_nodes(100)
```

```python
def add_random_edge(G):
    v1 = random.choice(list(G.nodes()))
    v2 = random.choice(list(G.nodes()))
    print(v1,v2)
    if v1!=v2:
        G.add_edge(v1,v2)
    return G
```

```python
H = add_random_edge(G)
```

```python
list(H.edges())
```

```python
#keeps adding random edges in G till it becomes connected
def add_till_connectivity(G):
    while(nx.is_connected(G)==False):
        G = add_random_edge(G)
    return G
```

```python
G = add_till_connectivity(G)
```

```python
G.edges()
```

```python
G.number_of_edges()
```

```python
nx.is_connected(G)
```

```python
#creates an instance of the entire process. It takes as input number of nodes and returns the number of edges required for connectivity
```

```python
def create_instance(n):
    G = add_nodes(n)
    G = add_till_connectivity(G)
    return G.number_of_edges()
```

```python
create_instance(10)
```

```python
create_instance(20)
```

```python
def plot_connectivity():
    x = []
    y = []
    i = 10 #i is no. of nodes
    while(i<=400):
        x.append(i)
        y.append(create_instance(i))
        i += 10
    plt.xlabel('Number of nodes')
    plt.ylabel('Number of edges required to connect the graph')
    plt.title('Emergence of connectivity')
    plt.plot(x,y)
    plt.show()
```

```python
plot_connectivity()
```
