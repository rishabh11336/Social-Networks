```python
def add_foci_nodes(G):
    n = G.number_of_nodes()
    i = n+1
    foci_nodes = ['gym', 'eatout', 'movie_club', 'karate_club', 'yoga_club']
    for j in range(0, 5):
        G.add_node(i)
        G.nodes[i]['name'] = foci_nodes[j]
        G.nodes[i]['type'] = 'foci'
        i = i+1
```

```python
def cmn(u, v, G):
    nu = set(list(G.neighbors(u)))
    nv = set(list(G.neighbors(v)))
    return len(nu & nv)
```

```python
def plot_density():
    x = []
    y = []
    for i in range(0,11):
        G = nx.read_gml('evolution_'+str(i)+'.gml')
        x.append(i)
        y.append(nx.density(G))
    plt.xlabel('Time')
    plt.ylabel('Density')
    plt.title('Change in Density')
    plt.plot(x,y)
    plt.show()
```

```python
def plot_obesity():
    x = []
    y = []
    for i in range(0,11):
        G = nx.read_gml('evolution_'+str(i)+'.gml')
        x.append(i)
        y.append(obesity(G))
    plt.xlabel('Time')
    plt.ylabel('Obesity')
    plt.title('Change in Obesity')
    plt.plot(x,y)
    plt.show()
```

```python
def obesity(G):
    num = 0
    for each in list(G.nodes()):
        if G.nodes[each]['name']==40:
            num += 1
    return num
```
